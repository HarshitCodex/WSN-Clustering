```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from scipy.spatial.distance import euclidean
fig, ax = plt.subplots()
```

```python
colors = ['b', 'orange', 'g', 'r', 'c', 'm', 'y', 'k', 'Brown', 'ForestGreen']
X,y = make_blobs(
    n_samples=200 ,n_features=2,
    centers=5,cluster_std=0.7,
    shuffle=True, random_state=1
)
import skfuzzy as fuzz
```

```python
alldata = np.vstack((X[:,0], X[:,1]))
fpcs = []
for i in range(1 , 10):
    cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(
        alldata, i, 2, error=0.005, maxiter=1000, init=None)
    cluster_membership = np.argmax(u, axis=0)
        
    plt.scatter(X[:,0],X[:,1],c=cluster_membership,cmap = 'rainbow')
    
    plt.xlabel("X-Value")
    plt.ylabel("Y-Value")
    plt.show()
    fpcs.append(fpc)
```

```python
fig2, ax2 = plt.subplots()
ax2.plot(range(1, 10), fpcs)
plt.xlabel('number of clusters')
plt.ylabel('fuzzy partition coefficient (FPC)')

plt.savefig('fuzzy_elbow_method.png')
```
